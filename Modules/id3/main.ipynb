```python
import pandas as pd

# Data
features = ['Player -1', 'Player +1', 'outcome']
games = pd.DataFrame(
    [
        ('Rock', 'Rock', 0),
        ('Rock', 'Paper', 1),
        ('Rock', 'Scissors', -1),
        ('Rock', 'Lizard', -1),
        ('Rock', 'Spock', 1),

        ('Paper', 'Rock', -1),
        ('Paper', 'Paper', 0),
        ('Paper', 'Scissors', 1),
        ('Paper', 'Lizard', 1),
        ('Paper', 'Spock', -1),

        ('Scissors', 'Rock', 1),
        ('Scissors', 'Paper', -1),
        ('Scissors', 'Scissors', 0),
        ('Scissors', 'Lizard', -1),
        ('Scissors', 'Spock', 1),

        ('Lizard', 'Rock', 1),
        ('Lizard', 'Paper', -1),
        ('Lizard', 'Scissors', 1),
        ('Lizard', 'Lizard', 0),
        ('Lizard', 'Spock', -1),

        ('Spock', 'Rock', -1),
        ('Spock', 'Paper', 1),
        ('Spock', 'Scissors', -1),
        ('Spock', 'Lizard', 1),
        ('Spock', 'Spock', 0),
    ],
    columns=features
)

# Dictionary for mapping
mapping = {'Rock': 0, 'Paper': 1, 'Scissors': 2, 'Lizard': 3, 'Spock': 4}

# Apply the mapping to the DataFrame
games_numeric = games.replace({'Player -1': mapping, 'Player +1': mapping})

print(games_numeric)

# Train the DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier

# Features and target
X = games_numeric[['Player -1', 'Player +1']]
y = games_numeric['outcome']

# Train the DecisionTreeClassifier
dt = DecisionTreeClassifier(criterion='entropy').fit(X, y)

# Example prediction
example = pd.DataFrame([[0, 1]], columns=['Player -1', 'Player +1'])  # 'Rock' vs 'Paper'
prediction = dt.predict(example)

# print(f'Prediction for Rock vs Paper: {prediction[0]}')



# Function to interact with the classifier
def predict_outcome(player1_choice, player2_choice):
    # Dictionary for mapping
    mapping = {'Rock': 0, 'Paper': 1, 'Scissors': 2, 'Lizard': 3, 'Spock': 4}

    # Check if inputs are valid
    if player1_choice not in mapping or player2_choice not in mapping:
        return "Invalid choices. Please choose from 'Rock', 'Paper', 'Scissors', 'Lizard', 'Spock'."

    # Map inputs to numerical values
    player1_numeric = mapping[player1_choice]
    player2_numeric = mapping[player2_choice]

    # Create a DataFrame for the example
    example = pd.DataFrame([[player1_numeric, player2_numeric]], columns=['Player -1', 'Player +1'])

    # Predict the outcome
    prediction = dt.predict(example)[0]

    # Map the prediction back to the outcome
    outcome_mapping = {0: "Draw", 1: "Player +1 wins", -1: "Player -1 wins"}
    return outcome_mapping[prediction]

# Example usage
player1_choice = input("Enter choice for Player -1 (Rock, Paper, Scissors, Lizard, Spock): ")
player2_choice = input("Enter choice for Player +1 (Rock, Paper, Scissors, Lizard, Spock): ")
result = predict_outcome(player1_choice, player2_choice)
print(f'Result: {result}')



```

```python
from sklearn.tree import DecisionTreeClassifier
import numpy as np

# Define the game rules and mapping for choices
rules = {
    1: {"beats": [3, 4], "loses_to": [2, 5]},
    2: {"beats": [1, 5], "loses_to": [3, 4]},
    3: {"beats": [2, 4], "loses_to": [1, 5]},
    4: {"beats": [2, 5], "loses_to": [2, 3]},
    5: {"beats": [1, 3], "loses_to": [2, 4]}
}

choices = {
    1: "Rock",
    2: "Paper",
    3: "Scissors",
    4: "Lizard",
    5: "Spock"
}

# Prepare the training data for the ID3 algorithm
X_train = []
y_train = []

for key, value in rules.items():
    for beats in value["beats"]:
        X_train.append([key, beats])
        y_train.append(1)  # 1 represents win
        
    for loses_to in value["loses_to"]:
        X_train.append([key, loses_to])
        y_train.append(0)  # 0 represents lose

X_train = np.array(X_train)
y_train = np.array(y_train)

# Train the Decision Tree Classifier (ID3 algorithm)
clf = DecisionTreeClassifier(criterion="entropy")
clf.fit(X_train, y_train)

def get_user_choice():
    while True:
        print("Enter your choice: ")
        print("1. Rock")
        print("2. Paper")
        print("3. Scissors")
        print("4. Lizard")
        print("5. Spock")
        choice = input("Enter the number corresponding to your choice: ")
        if choice in ['1', '2', '3', '4', '5']:
            return int(choice)
        else:
            print("Invalid input! Please enter a number between 1 and 5.")

def print_choices(user_choice, computer_choice):
    print(f"You chose: {choices[user_choice]}")
    print(f"Computer chose: {choices[computer_choice]}")

def determine_winner(user_choice, computer_choice):
    if user_choice == computer_choice:
        return "It's a tie!"

    # Determine the outcome using the decision tree
    outcome = clf.predict([[user_choice, computer_choice]])
    if outcome == 1:
        return "You win!"
    else:
        return "Computer wins!"

def play_game():
    print("Welcome to Rock Paper Scissors Lizard Spock!")
    
    while True:
        user_choice = get_user_choice()
        
        # Let the computer make a choice based on ID3 decision tree
        possible_choices = [1, 2, 3, 4, 5]
        computer_choice = None
        
        for choice in possible_choices:
            outcome = clf.predict([[user_choice, choice]])
            if outcome == 1:
                computer_choice = choice
                break
        
        print_choices(user_choice, computer_choice)
        
        result = determine_winner(user_choice, computer_choice)
        print(result)
        
        play_again = input("Do you want to play again? (yes/no): ").lower()
        if play_again != "yes":
            print("Thanks for playing!")
            break

# Start the game
play_game()
```
